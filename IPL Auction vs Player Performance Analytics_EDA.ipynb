# MODULE 1: IPL Auction Analytics

-  IPL Auction Analytics

```python
# Most Expensive Players
top_expensive = auction.sort_values(
    'SoldPrice',
    ascending=False
)

top_expensive[
    ['PlayerName','TeamName','AuctionYear','SoldPrice']
].head(20)
```

```python
#Visualization
plt.figure(figsize=(10,6))

sns.barplot(
    data=top_expensive.head(10),
    x='SoldPrice',
    y='PlayerName'
)

plt.title("Top 10 Most Expensive Players")
plt.show()
```

## Analysis 2


```python
# Team-wise Spending
team_spending = (
    auction.groupby('TeamName')['SoldPrice']
    .sum()
    .sort_values(ascending=False)
)

team_spending
```

```python
# Visualization
plt.figure(figsize=(12,6))

team_spending.plot(kind='bar')

plt.title("Total Spending by Team")
plt.ylabel("Amount")
plt.show()
```

## Analysis 3


```python
# Role-wise Spending
role_spending = (
    auction.groupby('Role')['SoldPrice']
    .sum()
    .sort_values(ascending=False)
)

role_spending
```

## Analysis 4
#

```python
#Nationality Spending
nationality_spending = (
    auction.groupby('Nationality')['SoldPrice']
    .sum()
    .sort_values(ascending=False)
)

nationality_spending
```

## Analysis 5


```python
# Capped vs Uncapped
capped_spending = (
    auction.groupby('CappedStatus')['SoldPrice']
    .sum()
)

capped_spending
```

## Analysis 6


```python
# Auction Trend by Year
yearly_spending = (
    auction.groupby('AuctionYear')['SoldPrice']
    .sum()
)
plt.figure(figsize=(12,6))

plt.plot(
    yearly_spending.index,
    yearly_spending.values,
    marker='o'
)

plt.title("Auction Spending Trend")
plt.show()
```

## MODULE 2: IPL Performance Analytics

-  IPL Performance Analytics

```python
# Top Batsmen
performance.sort_values(
    'Runs',
    ascending=False
)[['PlayerName','Runs']].head(20)
```

```python
# Top Bowlers
performance.sort_values(
    'Wickets',
    ascending=False
)[['PlayerName','Wickets']].head(20)
```

```python
#Best Strike Rate
performance[
    performance['Runs'] >= 500
].sort_values(
    'StrikeRate',
    ascending=False
).head(20)
```

```python
#Best Economy
performance[
    performance['Overs'] >= 100
].sort_values(
    'Economy'
).head(20)
```

```python
#All-Rounders
performance['AllRounderScore'] = (
    performance['Runs']*0.03
    +
    performance['Wickets']*20
)
```

```python
performance.sort_values(
    'AllRounderScore',
    ascending=False
).head(20)
```

```python
# Performance Ranking
performance.sort_values(
    'PerformanceScore',
    ascending=False
).head(20)
```

##  Key Business Insights

- Chennai Super Kings spent the highest amount in auctions.
-  All-rounders attract the largest auction investments.
- YS Chahal is the highest wicket-taking bowler.
- Virat Kohli remains the highest run scorer.

## Analysis 10: Team Preference by Role

## Analysis 11: Batting vs Bowling Contribution

```python
performance['BattingContribution'] = (
    performance['Runs'] /
    performance['Runs'].max()
)

performance['BowlingContribution'] = (
    performance['Wickets'] /
    performance['Wickets'].max()
)
```

```python
sns.scatterplot(
    data=performance,
    x='BattingContribution',
    y='BowlingContribution'
)

# Insight

#Identifies true all-rounders.
```
